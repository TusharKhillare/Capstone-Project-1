# Data Cleaning

**1-Data cleaning of Playstore data**

```python
# Row with inadequate data
data[data['Type']!='Free'][data[data['Type']!='Free']['Price']=='0']
```

```python
# Droping inadequate data row
data.drop(data[data['Type']!='Free'][data[data['Type']!='Free']['Price']=='0'].index, inplace=True)
```

```python
#Finding value mismatched row
data[data['Genres']=='February 11, 2018']
```

```python
# Index of mismatched row
data[data['Genres']=='February 11, 2018'].index
```

```python
#Droping row from dataframe
data.drop(data[data['Genres']=='February 11, 2018'].index, inplace=True)
```

```python
# Cleaning string function
def clean_it(num):
  """This function takes a string and replace the following characters if present, '+', ',' ,'$', 'M', 'k', 'NaN'"""
  if '+' in num:
    num = num.replace('+','')
  if ',' in num:
    num = num.replace(',','')
  if '$' in num:
    num = num.replace('$','')
  if 'M' in num:
    num = str(int(float(num.replace('M',''))*1000000))
  if 'k' in num:
    num = str(int(float(num.replace('k',''))*1000))
  if 'NaN' in num:
    num = '0'
  else:
    pass
  return num
```

```python
# Cleaning  undesirable characters 

#Changing'Reviews' into valid numeric values
data['Reviews'] = pd.to_numeric(data['Reviews'])

#Changing 'Size'of column values into valid numeric values
data['Size'] = data['Size'].apply(lambda x: str(x).replace('Varies with device', 'NaN') if 'Varies with device' in str(x) else x)
data['Size'] = pd.to_numeric(data['Size'].map(lambda x: clean_it(x)))

#Changing 'Installs' column values into valid numeric values
data['Installs'] = pd.to_numeric(data['Installs'].map(lambda x: clean_it(x)))

#Changing  'Price' column values into valid numeric values
data['Price'] = pd.to_numeric (data['Price'].map(lambda x:clean_it(x)))
```

# Data Preparation
Unique app details in a dataframe

```python
# Eliminating duplicate apps
duplicate_app = data['App'].duplicated().any()
duplicate_app
```

```python
# App Count
data['App'].value_counts()
```

```python
# Last rows of data for each app
last_row_data = data.groupby('App').tail(1).reset_index()
app_max_review = last_row_data.loc[last_row_data.groupby(['App'])['Reviews'].idxmax()]
```

```python
# Genres
top_genres = app_max_review.Genres.value_counts().reset_index().rename(columns={'Genres':'Count','index':'Genres'})
```

```python
top_genres
```

```python
app_max_review[app_max_review['Price'] == 0]
```

```python
# Free app install counts
genres_installs_free = app_max_review[app_max_review['Price'] == 0].groupby(['Genres'])[['Installs']].sum()
genres_installs_free
```

```python
# Paid app install counts
genres_installs_paid = data[data['Price']!= 0].groupby(['Genres'])[['Installs']].sum().rename(columns={'Installs':'Paid_app_installs'})
```

```python
# Mean Rating
genres_ratings = app_max_review.groupby(['Genres'])[['Rating']].mean()
genres_ratings
```

```python
# All the data previous dataframes
top_genres_installs = pd.merge(top_genres, genres_installs_free, on='Genres')
top_genres_installs = pd.merge(top_genres_installs, genres_installs_paid, on='Genres')
top_genres_ratings = pd.merge(top_genres_installs, genres_ratings, on='Genres')

#Getting top 50 data frames based on the Genres
top_50_genres = top_genres_ratings.head(50)
top_50_genres
```

**2-Top 10 installed apps in the Play Store in any category** 

```python
def findtop10incategory(str):
    str = str.upper()
    top10 = data[data['Category'] == str]
    top10apps = top10.sort_values(by='Installs', ascending=False).head(10)
    # Top_Apps_in_art_and_design
    plt.figure(figsize=(10,5))
    plt.title('Top 10 Installed Apps',size = 15);    
    graph = sns.barplot(x = top10apps.App, y = top10apps.Installs)
    graph.set_xticklabels(graph.get_xticklabels(), rotation= 45, horizontalalignment='right');
```
